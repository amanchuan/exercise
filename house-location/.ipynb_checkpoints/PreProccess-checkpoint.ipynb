```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
train_data = pd.read_csv("train_ccf_first_round_user_shop_behavior.csv") # # 存在很多记录是同一个用户，比如存在用户 u_7576987 有506条记录。
shop_inf = pd.read_csv("train_ccf_first_round_shop_info.csv")      # 共有97家商场，存在一个商场有很多商铺，比如商场 m_2224 有220家商铺。
test_data = pd.read_csv("ABtest_evaluation_public.csv")
```

```python
position_long = shop_inf['longitude']
position_lat = shop_inf['latitude']
plt.scatter(position_long,position_lat)
```

```python
train_data['longitude'][train_data.longitude<50] = 100
train_data['latitude'][train_data.latitude<20] = 20
user_position_x = train_data['longitude']
user_position_y = train_data['latitude']
plt.scatter(user_position_x,user_position_y)
```

```python
num_train = train_data.shape[0]
mall_id = []
for i in range(num_train):
    mall_id.append(shop_inf['mall_id'][shop_inf.shop_id == train_data['shop_id'][i]])
```

```python
dataframe = pd.DataFrame(mall_id)
dataframe.to_csv("shop_to_mall.csv", index=False)
print("***")
```
